```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import tellurium as te; te.setDefaultPlottingEngine('matplotlib')
import matplotlib.pyplot as plt
%matplotlib inline

model = '''

    X1=0;
    Y1=1;
    Z=0;
    Y=0;

    Y'= (X1/0.1)^2  /  (1+(X1/0.1)^2) - Y
    Z'= ((X1/0.1)^2 * (Y1/0.5)^2) /  ((1+(X1/0.1)^2) * (1+(Y1/0.5)^2)) - Z


 

    E1: at (time>0): X1= 0;
    E2: at (time>1): X1=1;
    E3: at (time>4): X1=0
'''

r = te.loada(model)
r.simulate(0,7, 100000)
r.plot()
```
